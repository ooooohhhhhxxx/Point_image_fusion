```python
import numpy as np
import struct
```

```python
bin_path = '/Users/jinxuanchen/Files_Local/Point_image_fusion/kittidata/odometry/sequences/00/velodyne/000010.bin'
outpath = '/Users/jinxuanchen/Files_Local/Point_image_fusion/kittidata/odometry/sequences/00/point_txt/000010.txt'
```

```python
pc_list = []
with open(bin_path, 'rb') as f:
        content = f.read()
        pc_iter = struct.iter_unpack('ffff', content)
        for idx, point in enumerate(pc_iter):
            pc_list.append([point[0], point[1], point[2],point[3]])
pc_list
```

```python
pc_list[:3]
```

```python
pc = np.array(pc_list)
pc[:,0:3]
```

```python
P0 = np.matrix([[7.188560000000e+02,0.000000000000e+00,6.071928000000e+02,0.000000000000e+00],
                [0.000000000000e+00,7.188560000000e+02,1.852157000000e+02,0.000000000000e+00],
                [0.000000000000e+00,0.000000000000e+00,1.000000000000e+00,0.000000000000e+00]])
P0 = np.insert(P0,3,values=[0,0,0,1],axis=0)
P0
```
